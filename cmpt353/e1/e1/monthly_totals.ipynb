```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('precipitation.csv', parse_dates=[2])
df
```

```python
def date_to_month(d):
    # You may need to modify this function, depending on your data types.
    return '%04i-%02i' % (d.year, d.month)
```

```python
df['month'] = df['date'].apply(date_to_month)
df
```

```python
grouped_total = df.groupby(['name','month']).sum().reset_index()
grouped_count = df.groupby(['name','month']).count().reset_index()
grouped_count
```

```python
totals = grouped_total.pivot(index='name', columns='month', values='precipitation')
counts = grouped_count.pivot(index='name', columns='month', values='precipitation')
```

```python
totals.to_csv('totals1.csv')
counts.to_csv('counts1.csv')
np.savez('monthdata1.npz', totals=totals.values, counts=counts.values)
```
